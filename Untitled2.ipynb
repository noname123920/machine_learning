```python
# 🔹 1. Проверка CUDA
import torch
if not torch.cuda.is_available():
    print("❌ CUDA не доступен! Включи GPU в Runtime → Change runtime type → GPU")
    exit()
print(f"✅ GPU: {torch.cuda.get_device_name(0)}")

# 🔹 2. Установка (если нужно)
!pip install diffusers transformers accelerate pillow -q

# 🔹 3. Импорт
from diffusers import AutoPipelineForImage2Image
from PIL import Image
from google.colab import files

# 🔹 4. Загрузка SD-Turbo
print("🚀 Загружаем SD-Turbo...")
pipe = AutoPipelineForImage2Image.from_pretrained(
    "stabilityai/sd-turbo",
    torch_dtype=torch.float16
)
pipe.to("cuda")
print("✅ Модель готова!")

# 🔹 5. Загрузка foto.png
print("📁 Загрузи foto.png...")
uploaded = files.upload()
image = Image.open("foto.png").convert("RGB")
image.thumbnail((512, 512))

# 🔹 6. JoJo промпт
prompt = "anime style, JoJo's Bizarre Adventure, bold outlines, dramatic shading, purple and gold colors, intense eyes, sharp features, high contrast, cel shaded, masterpiece"

# 🔹 7. Генерация
print("🎨 Генерируем...")
result = pipe(
    prompt=prompt,
    image=image,
    strength=0.8,
    guidance_scale=0.0,
    num_inference_steps=4
).images[0]

# 🔹 8. Скачать
display(result)
result.save("jojo_output.png")
files.download("jojo_output.png")
print("✨ Готово!")
```
